# Memantau Penggunaan Memory suatu Data Frame

Memanggil info()dengan menyertakan parameter memory_usagedengan nilai deep untuk menampilkan informasi detail dari memory pada data frame df_titanic.

# Memantau Penggunaan Memory untuk Setiap Kolom dari suatu Data Frame

Memanggil memory_usage dengan parameter deep yang diberi nilai True untuk menampilkan informasi penggunaan memory di setiap kolom pada data frame df_titanic.

```python
df_titanic.memory_usage(deep=True)
```
