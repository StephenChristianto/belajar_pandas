# persiapan data frame 

1.membuat data frame dengan jumlah barisnya 5 dan jumlah kolomnya 5 

2.memberi label untuk kolomnya 'ABCDE'

3.memanggil fungsi np.random.randint untuk mengisi data frame dengan jangkauan dari 1 sampai 20 dengan size sesuai baris dan kolom 

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                                     columns=cols)
df
```

memecah data frame menjadi dua bagian  

1.untuk pecahan pertama kita akan copy data frame dengan parameter deep yang bernilai True yang akan ditampung dengan df1

2.lalu kita membuang index 1 dan index 4

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])
df1
```

pemecahan bagian kedua 

1.kita copy data frame dengan parameter deep dengan nilai True yang akan ditampung di df2

2.lalu akan drop atau membuang index 0 dan 3

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

# Menggabungkan Dua Data Frame

1.disini kita akan menggunakan cara inner dan outer

2.lalu kita memanggil fungsi df_inner = pd.merge yang akan membutuhkan 2 parameter utama disini kita menggunakan df1 dan df 2 sebagai parameter 

3.disini kita juga bisa mengatur di how='inner' untuk merging 

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

1.disini kita memanggil fungsi df_outer = pd.merge yang akan membutuhkan 2 parameter utama disini kita menggunakan df1 dan df 2 sebagai parameter 

2.disini kita juga bisa mengatur di how='outer' untuk merging 

```python
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```
