# 24 - UTC dan konversi zona waktu pada Python Pandas 

# Import Modules

```python
import pandas as pd 

print(pd.__version__)
```

# Persiapan Pandas Series

```python
s = pd.Series(range(1591683521, 1592201921, 3600))
s = pd.to_datetime(s, unit='s')
s.head()
```

# Pengaturan Zona Waktu (Time Zone)

disini kita bisa menstandarisasi zona waktu dengan  s.dt.tz_localize('UTC')

```python
s = s.dt.tz_localize('UTC')
s.head()
```

lalu melakukan konversi dengan s.dt.tz_convert('Asia/Jakarta')

```python
s = s.dt.tz_convert('Asia/Jakarta')
s.head()
```

melakukan konversi dengan waktu Australia 

```python
s = s.dt.tz_convert('Australia/Hobart')
s.head()
```
