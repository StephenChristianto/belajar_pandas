# Membuat Data Frame

Data frame yang akan dibentuk oleh suatu dictionary data yang terdiri dari dua data (nama dan tempat_kelahiran).

Membuat data frame yang menggunakan dictionary data dengan memanggil pd.DataFrame(data)

```python
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'], 
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = pd.DataFrame(data)
df
```

# Memecah Nama Depan dan Nama Belakang

1.jika semisal kita ingin membuat dua kolom baru yang  bernama nama_depan dan nama_belakang

^ Membuat dua kolom baru dengan nama kolomnya yaitu nama_depan dan nama_belakang dengan menggunakan fungsi df[['nama_depan', 'nama_belakang']]

2.nama_depan dan nama_belakang akan menampung data dari fungsi str.split() yang akan memisahkan nama depan dan nama belakang dari kolom nama dengan menyertakan karakter spasi ' ' sebagai pemisahnya. lalu disertai dengan parameter dengan nilai expand=True agar output berupa mutliple columns

^ dengan memanggil df['nama'].str.split(' ', expand=True)

```python

df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

# Memecah Kota dan Provinsi Dari Tempat Kelahiran

semisal kita mau membentuk dua kolom baru yaitu kota dan provinsi 

1.Membuat dua kolom baru dengan nama kolomnya yaitu kolom dan provinsi kita bisa memangil fungsi df[['kota', 'provinsi']]

2.kota dan provinsi akan menampung data dari fungsi dari str.split() yang akan memisahkan kota dan provinsi dari tempat kelahiran dengan menyertakan parameter dengan karakter koma ',' sebagai pemisahnya. lalu ,akan disertai parameter dengan nilai expand=True agar dapat terbentuk multiple columns

^ dengan memanggil df['tempat_kelahiran'].str.split(',', expand=True)

```python
df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```
