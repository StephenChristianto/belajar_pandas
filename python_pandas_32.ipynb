# 32 - Plotting dari pandas data frame

# Import Modules

```python
import pandas as pd 
import numpy as np

print(pd.__version__)
print(np.__version__)
```

# Persiapan Data Frame

-membuat data frame dengan memiliki 40 baris dan 5 kolom 

-masing masing kolom diberi nama ABCDE

-dan isi data frame merupakan angka yang di acak dengan random.randint 

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,20,size=(n_rows,n_cols)),
                  columns=cols)
df.head()
```

# Line plot

```python
df.head()
```

menggunakan modul df.plot dengan parameter kind dengan nilai line

```python
df.plot(kind='line')
```

membuat plot dengan menyertakan A dan B 

```python
df[['A','B']].plot(kind='line')
```

# Bar Plot

```python
df.head()
```

menggunakan fungsi df.plot dengan parameter kind dengan nilai bar

```python
df.plot(kind='bar')
```

jika kita mau melihat bar A dan B kita bisa menyertakan A dan B 

```python
df[['A','B']].head().plot(kind='bar')
```

jika ingin membuat bar vertikal bisa menggunakan parameter kind dengan nilai barh

```python
df[['A','B']].head().plot(kind='barh')
```

# Area plot

```python
df.head()
```

jika ingin melihat area plot pada A dan B , menggunakan parameter kind dengan nilai area

```python
df[['A','B']].plot(kind='area')
```

```python
df[['A','B']].head().plot(kind='area')
```

# Box Plot

```python
df.head()
```

```python
df.plot(kind='box')
```

# Histogram

```python
df.head()
```

```python
df.plot(kind='hist')
```
