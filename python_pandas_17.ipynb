# Persiapan Data Frame

1.menyiapkan data frame yang terdiri dari baris yang isinya 365 dikali dengan 24  , kolom dengan jumlah 2 

2.memberi label setiap kolomnya dengan nama col 1 dan col 2

3.memanggil pd.DataFrame.random.randint untuk mengacak nomor dengan jangkauan 1 sampai 20 untuk mengisi data frame 

4.memanggil pd.util.testing.makeDateIndeks dengan jumlah kolom yang kita buat dengan frekuensi yang di atur H atau intervalnya per jam

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'cols2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df
```

# Resampling data dengan interval monthly

disini kita bisa mengatur ulang interval waktunya , semisal kita ingin mengubah menjadi bulan dengan cara :

1.memanggil df.resample dengan menyertakan parameter m dan menyertakan data yang akan di resampling (disini kita menggunakan data col 1)

2.lalu memanggil .sum() dan .to_frame() 

```python
df.resample('M')['col1'].sum().to_frame()
```

# Resampling data dengan interval daily

untuk mengubah atau mengatur ulang data frame dengan interval daily , kita bisa menggunakan cara :

1.memanggil df.resample dengan menyertakan parameter 'D' dan menyertakan data yang akan di resampling (disini kita menggunakan data col 1)

2.lalu memanggil .sum() dan .to_frame() 

```python
df.resample('D')['col1'].sum().to_frame()
```
