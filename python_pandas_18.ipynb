# Membentuk Data Frame dari Numpy Array

1.membentuk baris dan kolom dengan jumlah baris 5 dan jumlah kolom 3

2.membentuk numpy array dengan memangil np.random.randint dengan jangkauan 1 sampai 20 dengan size sesuai baris dan kolom 

3.lalu hasil array tersebut di tampung ke dalam arr

```python
n_rows = 5
n_cols = 3

arr = np.random.randint(1, 20, size=(n_rows, n_cols))
arr
```

disini kita bisa memanggil data array yang dibuat untuk dimasukan ke pd.DataFrame dengan cara :

1.memanggil pd.DataFrame dengan menyertakan array 

2.menyertakan parameter columns dengan parameter ini menyertakan label ABC

```python
pd.DataFrame(arr, columns=tuple('ABC'))
```
