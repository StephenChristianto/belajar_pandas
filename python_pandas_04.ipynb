# Membuat Data Frame Sederhana

^ disini kita akan membaut data frame dimana baris memiliki 5 dan kolom ada 2 

^ untuk isi data frame kita akan menggunakan np.random.randint dimana isi data tersebut mernupakan angka yang sudah di acak yang jangkauannya 1 sampai 20 yang sizenya sesuai kolom dan baris 

^ lalu kita memanggil df['bil_pecahan'].astype('float') untuk mengubah atau konversi bilangan pecahan menjadi bilangan yang float

^ lalu kita menggunakan df.index = pd.util.testing.makeDateIndex(n_rows, freq='H') yang bertujuan membuat data index baru yang menggunakan frekuensi hours / jam

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan,bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns = cols)
df{'bil_pecahan'} = df['bil_pecahan'].astype(float)

df.index = pd.util.testing.makeDateIndex(n_rows,freq ='H')
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
df.dtypes
```

# Memilih kolom yang bertipe data numerik

^ untuk memilih kolom bertipe nurmerik kita menggunakan script tersebut

```python
df.select_dtypes(include='number')
```

^ untuk memilih kolom bertipe float  kita menggunakan script tersebut

```python

df.select_dtypes(include='float')
```

^ untuk memilih kolom bertipe int kita menggunakan script tersebut

```python
df.select_dtypes(include='int')
```

^ untuk memilih kolom bertipe string atau object kita menggunakan script tersebut

```python
df.select_dtypes(include='object')
```

^ untuk memilih kolom bertipe  data time kita menggunakan script tersebut

```python
df.select_dtypes(include='datetime')
```

```python
^ untuk memilih kolom dengan kombinasi tipe data  kita menggunakan script tersebut
```

```python

df.select_dtypes(include=['number', 'object'])
```
