# Import Modules

Modules yang digunakan :

pandas as pd

```python
import pandas as pd

print(pd.__version__)
```

# Persiapan Data Frame

disiini kita tidak akan menggunakan angka lagi , tetapi untuk kolom kita ada 3 bagian yaitu day,month,year

dan untuk baris nya mereka terdiri dari masing masing data yang berkolerasi 

```python
data = {'day' :[1, 2, 10, 25, 12], 
              'month':[1, 2, 4, 5, 6],
              'year':[2000, 2001, 2010, 2015, 2020]}

df = pd.DataFrame(data)
df
```

# Membentuk kolom bertipe datetime 

untuk kolom barunya kita harus menuliskan namanya , contoh df['penanggalan']

lalu kita panggil pd.to_datetime(df[['day', 'month', 'year']]) sebagai kolom yang ingin kita sertakan 

```python
df['penanggalan'] =  pd.to_datetime(df[['day', 'month', 'year']])
df
```
