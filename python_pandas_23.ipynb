# 23 - Seleksi baris pada Data Frame dengan query

# Import Modules

```python
import pandas as pd 

print(pd.__version__)
```

# Persiapan Data Frame

1.membuat data frame dengan dictionary d yang terdiri dari kolom satu dan kolom dua

```python
d = {'kolom_satu':[1, 2, 3, 4, 5], 
     'kolom dua':[10, 20, 30, 40, 50]}
df = pd.DataFrame(d)
df
```

# Seleksi Baris dengan query()

untuk memakai fungsi ini kita memangging df.query() dengan memasukkan staitmentnya

```python
df.query('kolom_satu > 2')
```

```python
df.query('`kolom dua` > 30')
```
