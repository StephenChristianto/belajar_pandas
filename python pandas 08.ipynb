# Persiapan Data Frame

^ data frame saat ini akan ada dua buah data , yaitu  data A dan B

^ data yang kita pakai ini memiliki nilai yang sama ,

```python
data = { 'A':[15, 15, 18, np.nan, 12, ], 
               'B' :[5, 15, 18, np.nan, 12, ]}

df = pd.DataFrame(data)
df
```

# Memeriksa kesamaan dengan operator ==

^ jika ingin memeriksa kesamaan nilai kolom a dan kolom b kita bisa menggunakan operator ==

^ tetapi ada kelemahan dari operator == yaitu pada data NaN disini tidak bisa membandingkan dengan sempurna

```python
df[ 'A'] == df[ 'B' ] 
```

# Memeriksa kesamaan dengan operator equals

^ berikut alternatif jika kita tidak memakai operator == tetapi dengan equals

^ penulisannya yaitu df[ 'A'] .equals( df[ 'B' ]), tetapi jika di balik penulisannya akan sama saja

```python
df[ 'A'] .equals( df[ 'B' ])
```

# Memeriksa kesamaan kedua data frame

^ disini kita copy data frame semula yang kita buat , lalu ditampung di df1 dengan kondisi sama namun objeknya berbeda

^ lalu kita uji dengan fungsi / operator equals 

```python
df1 = df.copy(deep=True) 

df.equals(df1)
```

^ lalu kita coba menbandingkan dua data frame menggunakan operator ==

^ namun cara terbaik untuk membandingkan dua data frame yaitu dengan operator equals 

^ kalau oprator == itu bukan untuk membandingkan data , karena tujuannya juga beda sehingga hasil nya berbeda

```python
df == df1 
```
