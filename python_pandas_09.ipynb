# Persiapan Data Frame

^ disini kita akan membuat data frame dengan jumlah baris 10 dan kolom 5

^ lalu kita memberikan label kepada kolom ABCDE dengan memanggil fungsi tuple to string

^ untuk isi data nya kita menggunakan fungsi pd.DataFrame.random.randint yang dimana terdapat angka yang di acak dan secara random dengan jangkauan 1 sampai 20 dengan size sesuai baris dan kolom

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df
```

# Membagi Data Frame menjadi dua secara acak berdasarkan proposi tertentu

^ pertama kita harus cek data shape dari data frame tersebut 

```python
df.shape
```

^ disini kita akan membagi data shape tersebut menjadi dua bagian dengan proporsi nya 0.7

^ dengan proporsi 70% untuk data frame pertama , dan 30 % untuk data frame kedua

^ dengan cara kita menulis script df.sample yang di berikan parameter frac yang nilainya proposi , df sample ini akan mengambil baris secara acak dengan 70 %

^ untuk data frame ke dua , kita memanggil data frame utuh yang kita drop sesuai baris yang telah di ambil sebesar 70% lalu sisanya (30%)akan di tampung di df_2

```python
proposi = 0.7
df_1 = df.sample(frac=proposi)
df_2 = df.drop(df_1.index)

print(f'df_1 shape : {df_1.shape}')
print(f'df_2 shape : {df_2.shape}')
```
