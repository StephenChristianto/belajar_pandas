# Membuat Data Frame

^ Data Frame ini memiliki jumlah baris dan kolom masing - masing 5 , dan di namakan ABCDE

^Lalu kita menggunakan tuple untuk memberi label terhadap kolom tersebut 

^lalu kita menggunakan  pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                 columns=cols)
^  fungsi ini berguna untuk mengisi Data Frame dengan angka yang di acak dari jangkauan 1 smpai 10 dengan size sesuai dengan kolom dan baris

```python
n_rows = 5 
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                 columns=cols)
df
```

# Mengganti label untuk sebuah kolom pada Data Frame

^ Contoh nya kita ingin mengganti nama atau label C dengan kata hobi 

^ kita menggunakan fungsi df.rename(columns={ 'C' : 'Hobi'}) 

^ disini menggunakan fungsi atau method rename untuk menamai , lalu kita tambah dengan parameter columns, 

^ jangan lupa untuk menulis nama kolom yang ingin diubah , disini kita menulis C  lalu akan di ubah dengan kata Hobi

```python
df.rename(columns={ 'C' : 'Hobi'}) 
```

# Mengubah Nama Label untuk banyak kolom pada Data Frame

```python
^ Disini kita menggunakan fungsi yang sama ,  lalu menulis nama kolom yang akan di ubah dengan menuliskan nama baru untuk melabelkan kolom 
```

```python
df.rename(columns={ 'C' : 'alamat' , 'A' : 'tanggal' , 'B' : 'nomor' , 'D' : 'kota'})
```
