# Persiapan Data Frame

disini kita akan membuat data frame dengan jumlah baris 10 dan jumlah kolom 5

lalu kita memberi nama atau label untuk tiap kolomnya dengan fungsi tuple dengan nama ABCDE

untuk isi data nya kita menggunakan angka yang dipilih secara acak dari jangkauan 1 sampai 20 yang ukurannya sesuai dengan kolom dan baris

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                 columns=cols)
df
```

# Seleksi kolom dan baris menggunakan loc

untuk menggunakan fungsi loc , kita bisa menggunakan dua buah bagian yang dipisahkan oleh koma 

tidak lupa untuk awal dan akir menggunakan kurung siku 

bagian pertama untuk loc mengatur baris

sedangkan bagian kedua loc mengatur kolom 

```python
df.loc[[0,3,4], ['B','E']]
```

# Seleksi baris dengan kondisi

disini kita bisa menyertakan kondisi , dengan contoh [df['B']>10} disini artinya dari data frame yang angka barisnya di atas nilai 10

dan untuk bagian kedua kita hanya menuliskan bagian kolomnya saja 


```python
df.loc[df['B']>10, ['B','D','E']]
```

# Slicing Data Frame dengan loc

disini kita akan slicing untuk bagian pertama yaitu [0 :4] yang artinya indek dari baris 0 sampai index baris 4 

lalu kita untuk bagian kedua kita hanya perlu menuliskan nama kolomnya saja

```python
df.loc[0:4, 'B':'D']
```
