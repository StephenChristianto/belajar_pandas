# Membentuk Data Frame

1.membentuk baris dan kolom dengan jumlah barisnya 5 dan jumlah kolomnya 2

2.memberi nama kolom dengan nama  omset dan operasional

3.memanggil fungsi np.random.randint dengan jangkauan 1 sampai 20 unutk mengisi data frame dengan angka acak

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df
```

disini agar terlihat nyata kita akan mengkonversikan isi tabel omset dikali kan 100.000 dan operasional dikalikan 10.000

```python
df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] * 10_000
df
```

disini kita akan memodifikasi tabel di atas dengan menambahkan tabel tanggal 

1.kita akan generate dengan fungsi pd.util.testing.makeDateindex(n_rows, freq='D') dimana kita menyertakan baris dengan intervakbya day yang akan ditampung di df.index

2.karena kita enggak membentuk data index kita harus me reset index

3.setelah di reset kita akan menamai ulang kolom indexnya dengan nama index : tanggal

```python
df.index = pd.util.testing.makeDateIndex(n_rows,freq='D')
df = df.reset_index()
df = df.rename(columns={'index':'tanggal'})
df
```

# Melakukan formattin tampilan Data Frame

disini kita akan merubah tampilan tabel di atas agar dapat menambah Rp dan mengubah tanggal menjadi tanggal bulan tahun 

1.menyesuaikan key dengan nama kolomnya

2.memanggil df.style.format

```python
formatku = {'tanggal':'{:%d/%m/%y}',
           'operasional':'Rp {:.2f}',
           'omset':'Rp {:.2f}'}
laporan = df.style.format(formatku)
laporan
```

memeriksa tipe data

```python
type(laporan)
```

disini kita juga bisa menyembunyikan indexnya 

```python
laporan.hide_index()
```

kita juga bisa menyertakan judul dari data frame tersebut

```python
laporan.set_caption('Data Omset dan Operasional')
```

```python
kita juga bisa higlight nilai maks dan nilai min dari data kolom tersebut 
```

```python
laporan.highlight_min('omset',color='pink')
laporan.highlight_max('omset',color='lightgreen')

laporan.highlight_min('operasional',color='lightblue')
laporan.highlight_min('operasional',color='grey')
```
