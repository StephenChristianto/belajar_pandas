# Import Modules

Modules yang digunakan :

```python
import pandas as pd 

print(pd.__version__)
```

# Persiapan Data Frame

1.Menyiapkan 2 buah data frame yang berbeda , data frame 1 akan di tampung di df1 dan data frame kedua akan di tampung di df2

```python
d1 = {'col1':[1,2,3],
     'col2':[10,20,30]}
df1 = pd.DataFrame(d1)
df1
```

```python

d2 = {'col1':[4,5,6],
     'col2':[40,50,60]}
df2 = pd.DataFrame(d2)
df2
```

# Menggabungkan (merge) dua Data Frame secara berdampingan

untuk melakukan merging yang dibutuhkan yaitu 2 buah data frame dan menyesuaikan left index dan right index

```python
df = pd.merge(df1, df2, left_index=True, right_index=True)
df
```
