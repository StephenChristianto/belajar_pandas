# Persiapan Data Frame

^ Sebelumnya kita membuat Data Frame , dengan kolom dan baris berjumlah 5 

^ lalu kita menamakan nya abcde dengan menggunakan fungsi tuple

^ untuk isi dari Data Frame nya kita menggunakan np.random.randint dengan jangkauan 1 sampai 10 yang sizenya sesuai dengan baris dan kolom 

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                  columns=cols)
df
```

# Menyertakan preflix kolom

^ Jika kita menggunakan add_preflix kita bisa menambahkan label di depan alfabet yang tertera pada kolom yang sudah tadi kita buat

```python
df.add_prefix('kolom_')
```

# Menyertakan Suffix Kolom

^ suffix ini berguna untuk menambahkan nama sesudah huruf atau alfabet yang kita gunakan untuk menamai kolom 

```python
df.add_suffix('_field')
```
