# Persiapan Data Frame

1.membuat data frame dengan jumlah baris 10 dan jumlah kolom 1

2.memberi nama kolom dengan nama "usia"

3.untuk isi data frame ini kita menggunakan angka yang diacak dengan jangkauan 1 sampai 99  dengna size sesuai dengan besar kolom dan baris 

```python
n_rows = 10
n_cols = 1
cols =('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),
                 columns=cols)
df
```

# Pengelompokkan Nilai Numerik ke Dalam Beberapa Kategori Menggunakan cut()

1.menentukkan label usia yang kita gunakan , disini kita menggunakan anak,dewasa,manula

2.untuk usia anak itu jangkauannya 0 sampai 18 tahun , dewasa dari 18 sampai 65 tahun , untuk manula dari 65 sampai 99

3.membuat kolom baru dengan fungsi df['nama kolom'] , disini kita menggunakan df['kelompok usia']

4.untuk pemotongan usia ini kita menggunakan pd.cut(df['usia'])  dengan spot nya yaitu bins yang terdiri dari usia 0,18,65,99

```python
df['kelompok_usia'] = pd.cut(df['usia'],
                                                  bins = [0, 18, 65, 99],
                                                  labels = ['anak', 'dewasa', 'manula'])
df
```
