```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

# Persiapan Data Frame

^ Pertama kita membuat data frame dengan jumlah kolom 5 dan baris 10

^ lalu kita beri label abcde untuk setiap kolomnya

^  untuk isi data frame kita menggunakan angka yang telah di acak dengan jangkauan 1 sampai 5 dengan size sesuai kolom dan baris 

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                   columns=cols)
df
```

# Selection dengan operator logika |

^ disini kita akan menseleksi nilai a yang bernilai 1 atu nilai a nya yang 3 

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

# Selection dengan fungsi isin()

^ fungsi isin() ini juga sama persis dengan fungsi | atau or namun penulisannya yang berbeda 

```python

df[df['A'].isin([1, 3])]
```

# Mengenali Operator Negasi

^ fungsi ini untuk memilih nilai yang tidak sama dengan nilai 1 dan 3

```python
df[~df['A'].isin([1, 3])]
```
