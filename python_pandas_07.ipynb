# Mengimport modules 

Di sini kita mengimport modules pandas sebagai pd 

```python
import pandas as pd 

print(pd.__version__)
```

# Persiapkan Data Frame

^ Disini kita membuat data frame dengan mengandung missing value

^disini kita menggunakan modul pd.util.testing.makeMissingDataframe dan modul .reset_index. 
 
 ^ pd.util.testing.makeMissingDataframe berfungsi untuk membuat data dummy yang mengandung missing value
 
 ^.reset_index.  berfungsi untuk membentuk index baru 
 
 ^  lalu kita memakai modul df.head() untuk dapat memakai 5 data teratas saja

```python
df = pd.util.testing.makeMissingDataframe() .reset_index() 
df.head()
```

^ disini kita merubah nama index nya menjadi z

```python
df = df.rename(columns={ 'index' : 'Z' }) 
df.head()
```

^ lalu kita membuat copy dari tabel tersebut yang copyan nya akan di simpan di df_backup

^ menggunakan deep copy yang nilai dari parameter deep nya True

```python
df_backup = df.copy(deep=True)
```

# Menghapus Setiap Kolom yang Mengandung Missing Value

^ Disini kita menggunakan script df.dropna dengan parameter axis yang bernilai columns

^kenapa diberi nilai columns? karena kita ingin menghapus/drop  columns

```python
df = df.dropna(axis= 'columns' ) 
df.head()
```

# Menghapus setiap Baris yang Mengandung Missing Value

^ disini kita harus mengembalikan data frame kita dengan menggunakan df_backup.copy dengan parameter deep yang bernilai True

^ Karena kita sudah copy maka df yang akan kita gunakan sudah menampung data frame yang awal kita buat

^ lalu kita menggunakan script df = df.dropna(axis='rows' ) , beda nya dengan kolom yaitu , jika baris ini nilai dari parameter nya ( axis) diberi nilai rows

^ lalu kita menggunakan df.head() untuk menampilkan hasilnya

^ Disini kita bisa menggunakan 2 pendekatan , pertama menghapus kolom yang ada missing value, kedua menghapus baris yang ada missing value

```python
df = df_backup.copy(deep=True ) 
df = df.dropna(axis='rows' )
df.head()
```

# Presentase missing values untuk tiap kolom

^  pertama kita harus mengambil copy data frame semua dengan backup copy dengan parameter deep bernilai True

^ lalu kita menggunakan df.isna yang berfungsi untuk mengambil semua missing value yang ada di kolom

^ lalu kita menggunakan . mean yang berfungsi untuk menghitung rata rata dari missing value tersebut

```python
df = df_backup.copy(deep=True ) 
df.isna().mean() 
```

^ disini kita bisa melihat bahwa 

kolom Z memiliki 0 %

kolom A memiliki 6 %

kolom B memiliki 20%

kolom C memiliki 10%

kolom D memiliki 3 %

^ dan jenis data missing value tersebut ialah float

# Menghapus Setiap Kolom yang mengandung missing values berdasakan threshold

^ Disini sedikit beda dari pendekatan pertama , karena disini kita tidak menghapus kolom atau baris , melainkan melakukan pertimbangan dengan missing value

^jika missing value disini masih bisa di ampuni , kita tidak akan membuangnya

^jika missing value disini sudah kelas berat maka akan kita buang

^ Disini kita harus mengatur threshold atau ambang batasnya , disini kita menggunakan 0.9 atau 90 %

^ lalu kita memanggil fungsi len dari kolom kita  dan akan di kalikan dengan 90 % tersebut , untuk menghitung 90% dari tabel kita itu non missing value

^ lalu kita memanggil fungsi df.dopna dengan parameter tresh yang bernilai threshold, lalu menggunakan paramter axis dengan nilai columns

```python
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns' ) 
df.head()
```
