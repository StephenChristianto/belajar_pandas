# Mengenal groupby() dan fungsi agregasi

1.agregasi berfungsi sebagai fugnsi yang dikenakan terhadap suatu kelompok data dan berkaitan dekat dengan grouping atau groupby()

2.jika ingin mengelompokan data frame berdasarkan species kita bisa menggunakan ungsi agregasi yaitu count() untuk menghitung jumlah barisnya.



```python
df.groupby('Species')['PetalLengthCm'].count().to_frame()
```

jika ingin mengelompokan data frame berdasarkan species kita bisa menggunakan ungsi agregasi yaitu mean() untuk menghitung nilai rata-ratanya.

```python
df.groupby('Species')['PetalLengthCm'].mean().to_frame()
```

jika ingin mengelompokan data frame berdasarkan species kita bisa menggunakan ungsi agregasi yaitu median() untuk menghitung nilai mediannya.

```python
df.groupby('Species')['PetalLengthCm'].median().to_frame()
```
