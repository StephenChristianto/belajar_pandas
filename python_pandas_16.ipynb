# Membuat Data Frame

# Data Frame dengan Multiple Indexes dari hasil groupping

Melakukan pengelompokkan atau groupping data berdasarkan sex (female dan male) dan pclass (1, 2, dan 3).

Mengamati data survived dan mengambil aggregate valuenya dengan mean()

Pemanggilan to_frame() agar tampilan output menjadi rapih

```python
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()
```

kita bisa menata ulang dengan cara melakukan groupping berdasarkan sex dan pclass serta mengamati data survived dan mengambil aggregate valuenya dengan mean() . lalu, kita dapat memanggil fungsi unstack().

```python
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()
```
