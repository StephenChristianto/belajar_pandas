# Membuat Data Frame

^ Data Frame yang akan dibuat memiliki jumlah baris dan kolom masing - masing 5  yang diberi nama A,B,C,D,E
^ Jika ingin memberi label pada Data Frame maka kita menggunakan tuple terhadap String
^ Lalu kita menulis script atau kodedf = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),  columns=cols)
^ Fungsi dari script atas untuk mengisi Data Frame kita dengan angka random dengan jangkauan 1 sampai 10 , dengan size sesuai baris dan kolom

```python
n_rows = 5
n_cols = 5
cols = tuple( 'ABCDE' )

df = pd.DataFrame(np.random.randint(1, 11, size=(n_rows, n_cols)),
                 columns=cols)

df
```

# Membalik urutan kolom

^ jika ingin membalikkan kolom yang ada di contoh atas yaitu A, B,C,D,E . kita bisa menggunakan fungsi loc

^ df.loc[ : (bagian pertama), : :-1 (bagian kedua)] 

^ bagian pertama unutk mengatur baris, untuk menyertakan semua baris kita bisa menggunakan tanda : 

^lalu bagian kedua untuk mengatur kolom , untuk membalikkan kolom kita bisa menulis : (blank) : -1 (unutk membalikkan urutan)


```python
df.loc[ : , : :-1] 
```

# Membalik urutan baris 

^ jika ingin membalikkan urutan baris , kita bisa menulis script : (blank) : -1

^ jika ingin menyertakan seluruh kolom kita perlu menambahkan (: :-1, (blank) : )

^ Tetapi agak efesien penulisan script kita tidak perlu menulis script yang menyertakan kolom 

```python
df.loc[ : :-1]  
```

# Membalik Urutan Baris dan Melakukan Penyesuaian ulang Index

^ Jika ingin membalik urutan baris ,namun index 0 tetap berada di atas kita bisa menggunakan script di bawah ini

df.loc[ : , : :-1].reset_index(drop = True) 

^ pada awalnya kita perlu memanggil df.loc , lalu membalik baris nya 

^lalu kita harus menambah kode baru yaitu .reset_index, yang berfungsi untuk mengatur ulang indexnya

^lalu kita menyertakan parameternya yaitu (drop)dengan nilai Ture
 
 ^ sehingga kalau kita jalankan script tersebut , barisnya akan terbalik namun index 0 sampai 4 akan tetap 


```python
df.loc[ : , : :-1].reset_index(drop = True) 
```
